```python
import numpy as np
import matplotlib.pyplot as plt
import less
```

```python
x = np.linspace(0, 2*np.pi, 50)
y = np.sin(x)

chart = less.Chart(9,6)
chart.title('PLOT')
chart.plot(x=x, y=y)
chart.render()
```

```python
x = np.linspace(0, 2*np.pi, 50)
y = np.sin(x)

chart = less.Chart(9, 6)
chart.title('SCATTER')
chart.scatter(x=x, y=y)
chart.render()
```

```python
x = ['2010', '2011', '2012', '2013', '2014']
y = [4, 7, 12, 19, 28]

chart = less.Chart(9, 6)
chart.title('VBAR')
chart.vbar(x=x, y=y)
chart.render()
```

```python
y = ['2010', '2011', '2012', '2013', '2014']
x = [4, 7, 12, 19, 28]

chart = less.Chart(9, 6)
chart.title('HBAR')
chart.hbar(x=x, y=y)
chart.render()
```

```python
x = np.array([0, 1, 2]*100)
y = np.random.random(300)**2 + x/4

chart = less.Chart(6, 6)
chart.title('JITTER')
chart.jitter(x=x, y=y, x_width=0.5)
chart.render()
```

```python
categories = ['Apples', 'Bananas', 'Cherries']
values = [2, 3, 5]

chart = less.Chart(4, 6)
chart.title('CAKE')
chart.cake(categories=categories, values=values)
chart.render()
```
